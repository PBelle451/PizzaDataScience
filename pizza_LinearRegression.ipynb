```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pylab import rcParams
rcParams ['figure.figsize'] = 15,6
```

```python
df = pd.read_csv('/content/pizza.csv')
df.head()
```

```python
print(df.dtypes)
```

```python
pepperoni_flavor = df[df['name'] == 'pepperoni']
print(pepperoni_flavor)
```

```python
pepperoni_price = pepperoni_flavor['price']
print(pepperoni_price)
```

```python
# prompt: Create a chart with pepperoni_flavor as X and pepperoni_price as Y

plt.plot(pepperoni_flavor['X'], pepperoni_price)
plt.xlabel('Pepperoni Flavor')
plt.ylabel('Pepperoni Price')
plt.title('Pepperoni Flavor vs. Pepperoni Price')
plt.show()
```

```python
# prompt: Try to predict the price of the pepperoni

from sklearn.linear_model import LinearRegression

# Prepare the data
X = pepperoni_flavor['X'].values.reshape(-1, 1)  # Reshape to a 2D array
y = pepperoni_price.values

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict the price for a new pepperoni ID (let's say ID 11)
new_id = np.array([[11]])
predicted_price = model.predict(new_id)

print("Predicted price for pepperoni ID 11:", predicted_price[0])
```
